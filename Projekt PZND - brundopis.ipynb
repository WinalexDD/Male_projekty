```python
import numpy
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 1))

myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
print(r)
print(p)
print(std_err)
```

```python
from scipy import stats
import numpy

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x1 = stats.mode(speed)
x2 = numpy.mean(speed)
x3 = numpy.median(speed)
x4 = numpy.std(speed)
x5=x4**2
x6 = numpy.percentile(speed, 75)

print(x1, x2, x3, x4, x5, x6)
```

```python
from sklearn.linear_model import LinearRegression

linear_reg=LinearRegression()

x=df.population.values.reshape(-1,1)
y=df.suicide_number.values.reshape(-1,1)

linear_reg.fit(x,y)

b0 =linear_reg.intercept_
b1=linear_reg.coef_
print("b0:",b0)
print("b1:",b1)
print("Prediction 5M:",linear_reg.predict([[5000000]]))
print("Prediction 10M:",linear_reg.predict([[10000000]]))
print("Prediction 15M:",linear_reg.predict([[15000000]]))
print(linear_reg.score(x, y))
```

```python
from sklearn.metrics import r2_score
y_pred=linear_reg.predict(x)
r2_score(y, y_pred)
```
