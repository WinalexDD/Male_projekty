```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel(r"C:\Users\winal\Desktop\studyja\Zeszyt1.xlsx")
df=pd.DataFrame(data)
```

```python
def A(x,n,i,K):
    suma=0
    i=i/100
    v=1/(1+i)
    for k in range(n):
        kpx=df.lx[x+k]/df.lx[x]
        qxk=(df.lx[x+k]-df.lx[x+k+1])/df.lx[x+k]
        suma=suma+((v**(k-1))*kpx*qxk)
    suma=suma+(v**n*df.lx[x+n]/df.lx[x])
    return suma*K
def a(x,n,i):
    suma=0
    i=i/100
    v=1/(1+i)
    for k in range(n):
        kpx=df.lx[x+k]/df.lx[x]
        suma=suma+(v**k)*kpx
    return suma
```

```python
x=30
n=20
i=10
KWOTA=10000
P=round(A(x,n,i,KWOTA),2)/round(a(x,n,i),2)
print("P =", round(P,2))
```

```python
L1=[]
L2=[]
L3=[]
for k in range(n):
    L1.append(round(A(x+k,n-k,i,KWOTA),2))
    L2.append(round(a(x+k,n-k,i),2))
    L3.append(abs(round(round(A(x+k,n-k,i,KWOTA),2)-round(a(x+k,n-k,i),2)*P,2)))
```

```python
tabela=pd.DataFrame({'A': L1, 'a': L2, 'V':L3})
fig, ax = plt.subplots()
fig.patch.set_visible(False)
ax.axis('off')
tabelawykres=plt.table(cellText=tabela.values, cellLoc='center', rowLabels=tabela.index, colLabels=tabela.columns, loc='center')
plt.gcf().set_size_inches(5, 7)
plt.savefig("tabela.png")
```
